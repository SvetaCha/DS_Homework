## ABC - Анализ (ABC-analysis)

ABC-анализ

- Отсортировать в порядке убывания
- Общая сумма оборота
- % от суммы (cumsum)
- 20% - A, 20-80% -> B, <80% -> C

```python
import pandas as pd
```

```python
df = pd.read_excel('ABC - Аналіз.xlsx')
```

```python
df.head()
```

```python
df.shape
```

```python
# Отсортировать в порядке убывания

df.sort_values(by=['Продажі'])
df
```

```python
# Убираем первого клиента (так как он сильно выбивается от других)

df.drop(0, axis=0, inplace=True)
df
```

```python
# Общая сумма продаж (переменная)
all_sales = df['Продажі'].sum()


# Считаем Накопительную Сумму по каждому клиенту (в новом столбце)
df['Накопительная Сумма'] = df['Продажі'].cumsum()


# И определяем % от всех продаж для каждого клиента (в новом столбце)
df['Procent'] = df['Накопительная Сумма']/all_sales*100
df
```

```python
# %АВС

df['%ABC'] = df['Накопительная Сумма']/all_sales*100
df.head()
```

```python
# Определям категории для АВС теста

df.loc[df['Procent'] <= 20, 'Категория АВС'] = 'A'
df.loc[(df['Procent']>20)& (df['Procent']<= 80), 'Категория АВС'] = 'B'
df.loc[80 < df['Procent'] , 'Категория АВС'] = 'C'
df
```

```python
# Определям категории для АВС теста (через функцию)

def abc (p):
    if p <= 20:
        return 'A'
    elif p >20 and p <= 80:
        return 'B'
    else:
        return 'C'
          
df['ABC'] = df['%ABC'].map(abc)
df  
```

```python
from flask import Flask

app = Flask(__name__)

def parser(string):
    t = str(string)
    return t

@app.route('/<string>')
def category(string):
    print(string)
    t = parser(string)
    result = df[df['Клієнт'] == t].loc[:, 'ABC'].values[0]
    return f'Клієнт належить до групи {result}'

app.run()

```
